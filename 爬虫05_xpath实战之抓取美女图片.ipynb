```python
import requests
import os
from lxml import etree


dirname = 'GirlsImage'
if not os.path.exists(dirname) :
    os.mkdir(dirname)
    
# 如果要爬取多页，需要定义一个通用的url模板,不可变的
url0 = 'http://pic.netbian.com/4kmeinv/index_%d.html'
for page in range(1, 6) :
    if page == 1 :
        url = 'http://pic.netbian.com/4kmeinv/'
    else :
        url = url0%page
    
    
    headers = {
        'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'
    }
    response = requests.get(url = url, headers = headers)
    response.encoding = 'gbk'
    page_text = response.text
    tree = etree.HTML(page_text)
    lilist = tree.xpath('//div[@class = "slist"]/ul/li')
    for li in lilist :
        #print(type(li))
        title = li.xpath('./a/img/@alt')[0] + '.jpg' # 局部数据解析
        # 局部数据解析的时候，xpath表达式要使用./操作，表示当前的局部数据.
        img_src = 'http://pic.netbian.com/' + li.xpath('.//img/@src')[0]
        img_data = requests.get(url = img_src, headers = headers).content  # .content图片是byte类型数据
        imgPath = dirname + '/' + title
        with open(imgPath, 'wb') as p :
            p.write(img_data)
        print(title, img_src, '保存成功！')
print('好了可算搞完了！')
    
```
